```python
import random
import math
student_id=input()
student_id=student_id.replace('0','8')             #replace 0 with 8
#print(student_id)
min_point=int(student_id[4])
max_point=int(student_id[len(student_id)-1:len(student_id)-3:-1])*1.5  #reverse of last 2 digit*1.5
winning_points=int(student_id[len(student_id)-1:len(student_id)-3:-1])
#print(min_point,max_point,winning_points)
shuffle=int(student_id[3])                   #4th digit
#shuffle=8
random_point=[]
def alpha_beta(position,depth,isMaximize,alpha,beta):
  left=position[:len(position)//2]
  right=position[len(position)//2:]
  if depth==0:
    return position[0]
  if isMaximize:
      alpha=-math.inf
      #print(alpha)
      #print(position)
      #left=position[:len(position)//2]
      #right=position[len(position)//2:]
      for child in (left,right):
       # maxp=-math.inf
        # print(child)
         p=alpha_beta(child,depth-1,False,alpha,beta)
         alpha=max(p,alpha)
         if(alpha>=beta):
           break
      return alpha
  else:
      beta=math.inf
      #print(position)
      for child in (left,right):
       # maxp=-math.inf
         p=alpha_beta(child,depth-1,True,alpha,beta)
         beta=min(p,beta)
         if(alpha>=beta):
           break
      return beta

#Task1
for i in range(8):
  a=random.randint(min_point,max_point)
  random_point.append(a)
#print(random_point)
#random_point=[66,74,14,73,19,26,32,40]
#random_point=[36,26,112,57,85,80,107,28]
print("Task 1")
print("Generated 8 random points between the minimum and maximum point limits: ",random_point)
print("Total points to win "+str(winning_points))
alpha=-math.inf
beta=math.inf
a=alpha_beta(random_point,3,True,alpha,beta)                 #level=len(random_points)//2-1=8//2-1=4-1=3
print("Achieved point by alpha-beta pruning = ",a)
winner=""
if(a>=winning_points):
  winner="Optimus Prime"
else:
  winner="Megatron"
print("The winner is ",winner)
#Task2
print("Task 2")
output_values=[]
counter=0
for i in range(shuffle):
  random.shuffle(random_point)
  a=alpha_beta(random_point,3,True,alpha,beta)                 #level=len(random_points)//2-1=8//2-1=4-1=3
  output_values.append(a)
  if(a>=winning_points):
    counter+=1
#counter=0
print("After the shuffle :")
print("List of all points values from each shuffles: ",output_values)
print("The maximum value of all shuffle: ",max(output_values))
print("Won "+str(counter)+" times out of "+str(shuffle)+" shuffles")
#print(a)
```
